```python
import kagglehub

# Download selected version
path = kagglehub.model_download("google/universal-sentence-encoder/tensorFlow2/universal-sentence-encoder/1")

print("Path to model files:", path)
```

```python
embed = tf.saved_model.load(path)
```

```python
head = df.head(14_186).copy(deep=True)
tail = df.tail(14_186).copy(deep=True)
```

```python
head.lyrics = head.artist_name + "\n" + head.track_name + "\n" + head.lyrics
tail.lyrics = tail.artist_name + "\n" + tail.track_name + "\n" + tail.lyrics
```

```python
embedded_lyrics_head = embed(head.lyrics)
embedded_lyrics_tail = embed(tail.lyrics)
```

```python
vectors = pd.DataFrame(np.concatenate((tf.get_static_value(embedded_lyrics_head['outputs']),
tf.get_static_value(embedded_lyrics_tail['outputs']))))
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
data_to_fit_on = df.drop(['artist_name', 'track_name', 'lyrics'], axis=1, inplace=False)
data_to_fit_on.join(vectors)
```

```python
data_to_fit_on.info()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

label_encoder = LabelEncoder()
data_to_fit_on.genre = label_encoder.fit_transform(data_to_fit_on.genre)
data_to_fit_on.topic = label_encoder.fit_transform(data_to_fit_on.topic)
scaler = StandardScaler()
data_to_fit_on = scaler.fit_transform(data_to_fit_on)
data_to_fit_on
```

```python
new_df = pd.DataFrame(data_to_fit_on)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=1000, algorithm='ball_tree')
model.fit(data_to_fit_on)
```

```python
def flatten(lst):
    return [x for xs in lst for x in xs]

def recommend(df, transformed_df, model, list_ID):
    list_track = transformed_df.iloc[list_ID]
    neighbors = model.kneighbors(list_track, return_distance=False, n_neighbors=3)
    indexes = flatten(list(map(list, neighbors)))
    res = pd.DataFrame(df.iloc[indexes])[['artist_name', 'track_name', 'id']].values.tolist()
    return res
```

```python
result = recommend(df, new_df, model, list_ID=[0, 1, 2])
```
